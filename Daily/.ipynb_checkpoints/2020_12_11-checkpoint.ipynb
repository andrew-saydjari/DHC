```julia
#addprocs(2)

using FITSIO
using FFTW
using LinearAlgebra
using HDF5

@everywhere begin
    fink_filter_set  = fink_filter_bank(8,16);
    filter_list = fink_filter_list(fink_filter_set);

    function DHC_process_CF(file)
        println(file)
        f = FITS(file)
        data = read(f[1])
        close(f)
        data = convert(Array{Float64,2}, data)
        return speedy_DHC(data, filter_list)
    end
end

DHC_out_CF = pmap(DHC_process_CF, file_list)
DHC_out_CF = hcat(DHC_out_CF...)

h5write("DHC_out_CF.h5", "main/data", DHC_out_CF)
```

```julia
addprocs(2)

@everywhere using FITSIO
@everywhere using FFTW
@everywhere using LinearAlgebra
@everywhere using HDF5
@everywhere using Distributed

@everywhere push!(LOAD_PATH, pwd())
@everywhere using DHC_Utils

@everywhere begin
    fink_filter_set  = fink_filter_bank(8,16);
    filter_list = fink_filter_list(fink_filter_set);

    function DHC_process_CF(file)
        println(file)
        f = FITS(file)
        data = read(f[1])
        close(f)
        data = convert(Array{Float64,2}, data)
        return speedy_DHC(data, filter_list)
    end
end

DHC_out_CF = pmap(DHC_process_CF, file_list[1:10])
DHC_out_CF = hcat(DHC_out_CF...)

h5write("DHC_out_CF.h5", "main/data", DHC_out_CF)
```
